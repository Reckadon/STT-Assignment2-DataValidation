## Q.1 Implementing Validation with Pandera

- Validate the datatype of each column.
- Analyze each feature and write an appropriate check/scheme for them. For example, for “rideable_type” check the unique options and ensure that no other entry passes the check.

```python
import pandera as pa
import json

# define schema, validating the datatypes and writing some checks for each column
schema = pa.DataFrameSchema({
    "ride_id": pa.Column(str, checks=[pa.Check.str_length(min_value=16, max_value=16)], unique=True),  # ride ids are of len 16
    "rideable_type": pa.Column(pa.Category, checks=[pa.Check.isin(rideable_types)], coerce = True), # values must be from the rideable_types array only
    "started_at": pa.Column(pa.DateTime),
    "ended_at": pa.Column(pa.DateTime),
    "start_station_name": pa.Column(str, drop_invalid_rows=True),
    "start_station_id": pa.Column(str, drop_invalid_rows=True, coerce = True),  # force the station ids as strings, will throw error without coerce
    "end_station_name": pa.Column(str, drop_invalid_rows=True),
    "end_station_id": pa.Column(str, drop_invalid_rows=True, coerce = True),
    "start_lat": pa.Column(float, drop_invalid_rows=True),
    "start_lng": pa.Column(float, drop_invalid_rows=True),
    "end_lat": pa.Column(float, drop_invalid_rows=True),
    "end_lng": pa.Column(float, drop_invalid_rows=True),
    "member_casual": pa.Column(pa.Category, checks=[pa.Check.isin(["member", "casual"])], coerce=True) # values must be from the member_casual array only
  },
    # define checks at the DataFrameSchema-level
    checks=pa.Check(
        lambda df: df["ended_at"] > df["started_at"]
    )
, strict=True)

try:  # for manually validating
    schema.validate(df, lazy=True)
except pa.errors.SchemaErrors as e:
    print(json.dumps(e.message, indent=2))
```

since "start_station_name", "start_station_id", "end_station_name", "end_station_id", "end_lat", "end_lng" have null values, we can ignore/ make them nullable, or drop the rows

- Add data validation rules to verify that a ride's end time occurs after its start time using Pandera's decorator functionality.    

*Using pandera's decorator to add schema validation on a function to create a new column "ride_duration"*

```python
@pa.check_input(schema, lazy=True)
def calculate_ride_duration(dataframe: pd.DataFrame) -> pd.DataFrame:
    """function to make a ride_duration column, also checks for schema with pandera decorator"""
    dataframe["ride_duration"] = dataframe["ended_at"] - dataframe["started_at"]
    return dataframe

try:
    df_with_duration = calculate_ride_duration(df)
except pa.errors.SchemaErrors as e:
    print(json.dumps(e.message, indent=2))

df_with_duration
```

## Q.2 Great Expectations Suite:

```python
# Import required modules from GX library.
import great_expectations as gx

from great_expectations.checkpoint import (
    EmailAction
)
```

```python
# Create Data Context.
context = gx.get_context()

# Import sample data into Pandas DataFrame.
df = df

# Connect to data.
# Create Data Source, Data Asset, Batch Definition, and Batch.
data_source = context.data_sources.add_pandas("pandas")
data_asset = data_source.add_dataframe_asset(name="pd dataframe asset")

batch_definition = data_asset.add_batch_definition_whole_dataframe("batch definition")
batch_parameters = {"dataframe": df}
batch = batch_definition.get_batch(batch_parameters=batch_parameters)
print(batch.head(3))
```

- Validate the datatype of each column by creating an expectation suite.

```python
# Create an Expectation Suite
suite_name = "datatype_expectation_suite"
suite = gx.ExpectationSuite(name=suite_name)

# Add the Expectation Suite to the Data Context
suite = context.suites.add(suite)

# Create Expectations

expectation_0 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="ride_id", type_="str"
)

expectation_1 = gx.expectations.ExpectColumnValuesToBeInSet(
    column="rideable_type",
    value_set=rideable_types,
    mostly=.1
)

expectation_2 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="started_at", type_="datetime64"
)

expectation_3 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="ended_at", type_="datetime64"
)

expectation_4 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="start_station_name", type_="str"
)

expectation_5 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="start_station_id", type_="str"
)

expectation_6 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="end_station_name", type_="float64"
)

expectation_7 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="end_station_id", type_="str"
)

expectation_8 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="start_lat", type_="float64"
)

expectation_9 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="start_lng", type_="float64"
)

expectation_10 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="end_lat", type_="float64"
)

expectation_11 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="end_lng", type_="float64"
)

expectation_12 = gx.expectations.ExpectColumnValuesToBeInSet(
    column="member_casual",
    value_set=["member", "casual"],
    mostly=.1
)

# Validate Suite using Expectations

suite.add_expectation(expectation_0)
suite.add_expectation(expectation_1)
suite.add_expectation(expectation_2)
suite.add_expectation(expectation_3)
suite.add_expectation(expectation_4)
suite.add_expectation(expectation_5)
suite.add_expectation(expectation_6)
suite.add_expectation(expectation_7)
suite.add_expectation(expectation_8)
suite.add_expectation(expectation_9)
suite.add_expectation(expectation_10)
suite.add_expectation(expectation_11)
suite.add_expectation(expectation_12)

```

```python
# Validation Definations
definition_name = "datatype_validation"
validation_definition = gx.ValidationDefinition(
    data=batch_definition, suite=suite, name=definition_name
)

validation_definition = context.validation_definitions.add(validation_definition)
validation_definition_name = "datatype_validation"
validation_definition = context.validation_definitions.get(validation_definition_name)

# Validation Results
validation_results = validation_definition.run(batch_parameters=batch_parameters)
print(validation_results)

validation_definitions = [
    context.validation_definitions.get("datatype_validation")
]
```
